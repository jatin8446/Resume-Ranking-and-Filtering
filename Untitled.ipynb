```python
import PyPDF2

def extract_text(file):
    text = ""
    reader = PyPDF2.PdfReader(file)
    
    for page in reader.pages:
        content = page.extract_text()
        if content:
            text += content
    
    return text

file = open("resumes/resume.pdf", "rb")

text = extract_text(file)

print(text[:500])
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
job_desc = """
Looking for a Data Analyst with skills in Python, SQL, Power BI, Excel.
Knowledge of machine learning is a plus.
"""
```

```python
def calculate_score(resume_text, job_desc):
    documents = [resume_text, job_desc]
    
    tfidf = TfidfVectorizer(stop_words='english')
    tfidf_matrix = tfidf.fit_transform(documents)
    
    score = cosine_similarity(tfidf_matrix[0:1], tfidf_matrix[1:2])
    
    return score[0][0]
```

```python
score = calculate_score(text, job_desc)

print("Resume Score:", score)
```

```python
def extract_skills(job_desc):
    job_desc = job_desc.lower()
    
    common_skills = [
        "python","sql","excel","power bi","machine learning",
        "aws","azure","gcp","cloud","docker","kubernetes",
        "java","c++","javascript","react","node","django",
        "testing","selenium","manual testing","automation",
        "citrix","linux","windows","devops","git","tableau"
    ]
    
    detected = [skill for skill in common_skills if skill in job_desc]
    
    return detected
```

```python
def match_skills(resume_text, skills):
    resume_text = resume_text.lower()
    
    matched = [skill for skill in skills if skill in resume_text]
    
    return matched
```

```python
job_desc = "python sql azure"

skills = extract_skills(job_desc)

matched_skills = match_skills(text, skills)

print("Matched Skills:", matched_skills)
print("Total Match:", len(matched_skills), "/", len(skills))
```

```python
match_percent = (len(matched_skills) / len(skills)) * 100

print("Match Percentage:", match_percent, "%")
```

```python
if match_percent >= 60:
    print("Status: SELECTED")
else:
    print("Status: REJECTED")
```

```python
print("----- RESUME ANALYSIS -----")
print("Score:", score)
print("Matched Skills:", matched_skills)
print("Match %:", match_percent)

if match_percent >= 60:
    print("Final Result: SELECTED ")
else:
    print("Final Result: REJECTED ")
```
